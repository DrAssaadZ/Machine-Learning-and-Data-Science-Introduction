```python
import numpy as np
x = np.linspace(0,7,10)
```

```python
y = x**2 
```

**there are 2 methods to use matplotlib, a functional method and a OOP method**

## Functional method!

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('my first plot')
```

**we can display more than one plot at once using the function subplot by defining the number of rows and columns, and the index of the current plot**
> here we have an example of 1 line 2 columns

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object Oriented Method

```python
# the next 2 lines should always be in the same cell
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('second plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

## OO Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_xlabel('x1 label')
axes[1].set_xlabel('x2 label')

# fix layout appearance
plt.tight_layout()
```

## Size of figures

```python
# we add the parameter figsize and dpi
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4), dpi=80)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## saving figures

```python
fig.savefig('my_pic.png', dpi=100)
```

## legends

**to add a legend, we should add a parameter called label in every plot in the figure, and add the line "axes.legend()"**

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,'r', label='x^2')
axes.plot(x,x**3,'g', label='x^3')

axes.legend(loc=2)   # use the tab button to get information on the loc parameter

plt.tight_layout()
```

## Customizing the layout!

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#151697', linewidth=2, alpha=0.7, linestyle=':', label='yo', marker='o',
       markersize=12, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='red')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

#limitations
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
