```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

**from previous plot we can see that there are a certain correlation between the tip and the bill*

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex', color='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg', color='yellow')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde', color='black')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```
