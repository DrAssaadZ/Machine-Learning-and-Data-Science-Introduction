# K-means

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(200,2,4,1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='rainbow')
```

```python
sns.scatterplot(data[0][:,0],data[0][:,1], hue=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(4)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

in real world scenarios, we usually wont have labels for the data, but in our case we have labels and with it we will calculate the accuracy of this model

```python
fig ,(ax1,ax2) = plt.subplots(1,2,figsize=(10,7))
ax1.set_title('K means')
ax2.set_title('original')
ax1.scatter(data[0][:,0],data[0][:,1],c=km.labels_, cmap='rainbow')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='rainbow')
```
